## Silhouettes

```python
from rembg import remove 
from PIL import Image

## Path for input and output image
input_img = 'bicycle_1.png'
output_img = 'bicycle_out.png'

## loading and removing background
inp = Image.open(input_img)
output = remove(inp)

## Saving background removed image to same location as input image
output.save(output_img)
```

## None More Black

```python
from PIL import Image

# Load image
im = Image.open('bicycle_out.png')

# Extract the alpha channel and threshold it at 200
alpha = im.getchannel('A')
alphaThresh = alpha.point(lambda p: 255 if p>200 else 0)

# Make a new completely black image same size as original 
res = Image.new('RGB', im.size)

# Copy across the alpha channel from original
res.putalpha(alphaThresh)
res.save('bicycle_sil.png')
```

## Word Cloud

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
```

```python
stopwords = set(STOPWORDS)
```

```python
kanga_wc = WordCloud()
kanga_wc.generate(data)
```

```python
bicycle_mask = np.array(Image.open('assets/bicycle_sil.png'))
```

```python
fig = plt.figure(figsize=(2,3))
plt.imshow(bicycle_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
bicycle_wc = WordCloud(background_color='black', max_words=5000, mask=bicycle_mask, colormap = "cool", stopwords=stopwords)
bicycle_wc.generate(data)
fig = plt.figure(figsize=(21,19))
plt.imshow(bicycle_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
